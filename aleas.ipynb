```julia
using Distributions
```

```julia
rexp = rand(Exponential(1), 10^4);
y = rand(10^4);
y1 = -log.(y);
p1 = histogram(rexp, label="rexp")
p2 = histogram(y1, label="y1", color=:purple)
plot(p1,p2,layout=(1,2))
```

```julia
U = rand(3, 10^4);
X = -log.(U);
X = 2 .* sum(X, dims=1);
X = vec(X);
rchisq=rand(Chisq(6),10^4);
p1 = histogram(rchisq, label="rchisq(6)")
p2 = histogram(X, label="X", color=:purple)
plot(p1,p2,layout=(1,2))
```
